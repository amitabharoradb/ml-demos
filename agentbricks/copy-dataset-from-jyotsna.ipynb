```python
src_catalog = 'jyotsnacatalog'
src_schema = 'agentbricks'
src_volumes = ['calltranscripts', 'invoices', 'knowledgebase']
```

```python
target_catalog = 'amitabh_arora_catalog'
target_schema = 'demo_agentbricks_src'
target_volumes = ['calltranscripts', 'invoices', 'knowledgebase']
```

```python
spark.sql(f"CREATE CATALOG IF NOT EXISTS {target_catalog}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {target_catalog}.{target_schema}")
```

```python
for volume in target_volumes:
    spark.sql(f"""
        CREATE VOLUME IF NOT EXISTS {target_catalog}.{target_schema}.{volume}
    """)
```

```python
for volume in src_volumes:
    src_path = f"/Volumes/{src_catalog}/{src_schema}/{volume}/"
    tgt_path = f"/Volumes/{target_catalog}/{target_schema}/{volume}/"
    dbutils.fs.cp(src_path, tgt_path, recurse=True)
```

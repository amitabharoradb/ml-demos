```sql
WITH query_results AS (
  SELECT `text` AS input,
    ai_query(
      'kie-39912b26-endpoint',
      input,
      failOnError => false
    ) AS response
  FROM (
    SELECT `text`
       FROM `amitabh_arora_catalog`.`demo_agentbricks_tgt`.`invoices_parsed`
    LIMIT 20
  )
)
SELECT
  input,
  response.result AS response,
  response.errorMessage AS error
FROM query_results
```

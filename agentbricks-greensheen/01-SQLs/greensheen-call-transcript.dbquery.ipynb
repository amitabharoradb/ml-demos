```sql
WITH query_results AS (
  SELECT
    text AS input,
    ai_query(
      't2t-566b359a-endpoint',
      input,
      failOnError => false
    ) AS response
  FROM (
    SELECT text
    FROM `amitabh_arora_catalog`.`demo_agentbricks_tgt`.`calltranscripts_parsed`
    LIMIT 20
  )
)
SELECT
  input,
  response.result AS response,
  response.errorMessage AS error
FROM query_results
```
